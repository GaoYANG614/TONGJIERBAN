```python
"""
4.利率
"""
def futureInvestmentValue(benjin,lixi,years):
    i=years
    u=benjin
    p=lixi
    for i in range(1,i+1):
        l=u+u*p*i
        print(i,"  ",l)
futureInvestmentValue(10000,0.05/12,30)
```
