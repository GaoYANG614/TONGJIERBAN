```python
"""
5、
"""
class GA(object):
    def __init__(self,a,b,c,d,e,f):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
        self.e=e
        self.f=f
        self.x=0
        self.y=0
        self.z=0
    def isSolvable(self):
        z=self.a*self.d-self.b*self.c
        if z !=0:
            self.z=True
        else:
            self.z=False
    def get(self):
        self.x=(self.e*self.d-self.b*self.f)/(self.a*self.d-self.b*self.c)
        self.y=(self.a*self.f-self.e*self.c)/(self.a*self.d-self.b*self.c)
    def getX(self):
        if self.z==True:
            self.get()
            print(self.x)
        else:
            pass
    def getY(self):
        if self.z==True:
            self.get()
            print(self.y)
        else:
            pass
        
dd=GA(78,45,113,79,46,5)
dd.isSolvable()
dd.getX()
dd.getY()
```
